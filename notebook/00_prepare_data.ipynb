```python
import pandas as pd
```

```python
seed=42

chunks =pd.read_csv("D:/VsCodeProjects/Ecommerce_User_Analysis/data/UserBehavior.csv",names=['user_id','item_id','category_id','behavior_type','timestamp'],chunksize=1000000)

sample=[]

for chunk in chunks:
    sample.append(chunk.sample(frac=0.02,random_state=seed))

df=pd.concat(sample)
df.to_csv('D:/VsCodeProjects/Ecommerce_User_Analysis/data/sample.csv',index=False)
```
