# 时间戳转换

```python
df["datetime"] = pd.to_datetime(df["timestamp"], unit="s") + pd.Timedelta(hours=8)
```

# 提取时间特征

```python
df["date"] = df["datetime"].dt.date
```

```python
df["hour"] = df["datetime"].dt.hour
```

```python
df["weekday"] = df["datetime"].dt.day_name()
```

## 数据清洗结论

1. 基于固定随机种子对原始数据进行 2% 抽样，共获得 2,003,016 条用户行为记录；
2. 数据不存在缺失值，完整性较高；
3. 数据不存在完全重复记录，无需去重；
4. 将 Unix 时间戳转换为标准时间格式，并提取日期、小时及星期等时间特征；
5. 数据覆盖 2017 年 11 月 25 日至 2017 年 12 月 3 日的用户行为，可支撑后续用户行为分析、漏斗分析及 RFM 用户价值分析。
### 时间字段处理说明

原始数据中的 Unix 时间戳采用 UTC 标准时间记录。考虑到淘宝平台用户主要来自中国大陆，本项目将时间统一转换为 UTC+8（北京时间），并基于修正后的时间提取小时、日期及星期等特征，以保证后续行为分析结果符合实际业务场景。
